```python
file_path = 'Scaled Parameters.xlsx'
df = pd.read_excel('Scaled Parameters.xlsx')
```

```python
df = df[df['Parameter'] != 'Housing Material']

df = df[df['Parameter'] != 'Dimensions (Cubic mm)']

df = df[df['Parameter'] != 'Communication Range (RS-232)']

df = df[df['Parameter'] != 'Communication Range (RS-485)']

# Display the updated DataFrame
print(df.head())
```

```python
df
```

```python
# Set up the figure and plot
plt.figure(figsize=(19,7))

# Plot the two columns against 'Parameter'
plt.plot(df['Parameter'], df['Sulzer Equipment Controller EC 531'], label='Sulzer Equipment Controller EC 531', marker='o')
plt.plot(df['Parameter'], df['KirloSmart 1.1'], label='KirloSmart 1.1', marker='s')

# Add labels and title
plt.xlabel('Parameter')
plt.ylabel('Value')
plt.title('Comparison of Sulzer EC 531 vs KirloSmart 1.1')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Set up the figure and plot
plt.figure(figsize=(22,10))

# Set the position for the bars
bar_width = 0.35
index = range(len(df['Parameter']))

# Plot bars for each controller
plt.bar(index, df['Sulzer Equipment Controller EC 531'], bar_width, label='Sulzer EC 531')
plt.bar([i + bar_width for i in index], df['KirloSmart 1.1'], bar_width, label='KirloSmart 1.1')

# Add labels and title
plt.xlabel('Parameter')
plt.ylabel('Value')
plt.title('Comparison of Sulzer EC 531 vs KirloSmart 1.1')
plt.xticks([i + bar_width/2 for i in index], df['Parameter'], rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Set up the figure and plot with a larger size
plt.figure(figsize=(25,10))

# Set the position for the bars
bar_width = 0.35
index = range(len(df['Parameter']))

# Plot bars for each controller
plt.bar(index, df['Sulzer Equipment Controller EC 531'], bar_width, label='Sulzer EC 531')
plt.bar([i + bar_width for i in index], df['KirloSmart 1.1'], bar_width, label='KirloSmart 1.1')

# Add labels and title
plt.xlabel('Parameter', fontsize=16)  # Increased font size for x-axis label
plt.ylabel('Value', fontsize=16)      # Increased font size for y-axis label
plt.title('Comparison of Sulzer EC 531 vs KirloSmart 1.1', fontsize=22, fontweight='bold')  # Larger and bold title

# Set x-axis labels with larger size and bold text
plt.xticks([i + bar_width/2 for i in index], df['Parameter'], rotation=45, fontsize=14, fontweight='bold', color='black')

# Add a legend with a larger font size
plt.legend(fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```
